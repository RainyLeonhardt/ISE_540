```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.pipeline import make_pipeline
from imblearn.over_sampling import SMOTE
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

# Suppress specific FutureWarnings from sklearn
warnings.filterwarnings('ignore')

# Reading the data
data = pd.read_parquet('Labeled_data_100_3.Parquet')
data, data_t = data[:split], data[split:]

# Feature extraction
tfidf_expanded = pd.DataFrame(data['TF-IDF_Vector'].tolist())
word2vec_expanded = pd.DataFrame(data['Word2Vec_Vector'].tolist())
encoder = OneHotEncoder(handle_unknown='ignore')
categorical_data = encoder.fit_transform(data[['Product', 'Issue', 'Tags']]).toarray()
categorical_data_df = pd.DataFrame(categorical_data, columns=encoder.get_feature_names_out())
features = pd.concat([tfidf_expanded, categorical_data_df], axis=1)
target = data['Severity']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=42)

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Preprocessing validation data
tfidf_expanded_t = pd.DataFrame(data_t['TF-IDF_Vector'].tolist())
word2vec_expanded_t = pd.DataFrame(data_t['Word2Vec_Vector'].tolist())
categorical_data_t = encoder.transform(data_t[['Product', 'Issue', 'Tags']]).toarray()
categorical_data_df_t = pd.DataFrame(categorical_data_t, columns=encoder.get_feature_names_out())
features_t = pd.concat([tfidf_expanded_t, categorical_data_df_t], axis=1)
X_val = features_t
y_val = data_t['Severity']

# Initialize models
models = {
    'LogisticRegression': LogisticRegression(max_iter=100, random_state=42),
    'RandomForest': RandomForestClassifier(random_state=42),
    'GradientBoosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    'NeuralNetwork': MLPClassifier(random_state=42)
}

# Training and evaluating each model
for name, model in models.items():
    pipeline = make_pipeline(StandardScaler(), model)
    pipeline.fit(X_train_smote, y_train_smote)
    y_pred_test = pipeline.predict(X_test)
    y_pred_val = pipeline.predict(X_val)
    cv_scores = cross_val_score(model, X_train_smote, y_train_smote, cv=10)

    print(f"Model: {name}")
    print("Cross-validation scores:", cv_scores)

    # Compute the mean and standard deviation of the cross-validation scores
    print("Mean cross-validation score:", cv_scores.mean())
    print("Standard deviation of cross-validation scores:", cv_scores.std())

    print("Validation Set Accuracy:", accuracy_score(y_test, y_pred_test))
    
    # Plotting Confusion Matrix for Test Set
    cm = confusion_matrix(y_test, y_pred_test)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {name} - Validation Set')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

    # Print classification report for Test Set
    print("Validation Set Classification Report:\n", classification_report(y_test, y_pred_test))
    
    print("Test Set Accuracy:", accuracy_score(y_val, y_pred_val))
    # Plotting Confusion Matrix for Validation Set
    cm_val = confusion_matrix(y_val, y_pred_val)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_val, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {name} - Test Set')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

    # Print classification report for Validation Set
    print("Test Set Classification Report:\n", classification_report(y_val, y_pred_val))
    print("-" * 50)
```
